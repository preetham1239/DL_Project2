```python
from __future__ import print_function, division

import pandas as pd
from PIL import Image
import numpy as np
from torchvision import transforms
import matplotlib.pyplot as plt
import os
plt.ion()
```

```python
path_var = "test"
```

```python
p = transforms.Compose([transforms.Scale((640, 640))])
temp = 0
file_images = os.listdir('/WAVE/projects/COEN-342-Wi22/pr2/'+path_var+'/images/')
for i in file_images:
    img_path = '/WAVE/projects/COEN-342-Wi22/pr2/'+path_var+'/images/'+i
    img = Image.open(img_path)
    temp_image = p(img)
    print("Image ID is ", temp)
    temp_image.save('./dataset/'+path_var+'/images/'+i, format='JPEG')
    temp += 1
```

```python
df = pd.read_csv('/WAVE/projects/COEN-342-Wi22/pr2/'+path_var+'/labels.txt', sep=' ', names=['ID', 'class', 'cx', 'cy', 'width', 'height'])
df['cx'] = (df['cx']/960)
df['cy'] = (df['cy']/540)
df['width'] = (df['width']/960)
df['height'] = (df['height']/540)
df['ID'] = df['ID'].astype(np.int64)
df['class'] = df['class']-1
# df
```
